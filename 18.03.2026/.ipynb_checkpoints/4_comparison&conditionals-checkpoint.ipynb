- <b> Check if number 10 is present in the array. Accordingly, print "10 is Present" else "10 not Present"

```python
numbers_array = [1, 5, 10, 15, 20]
if 10 in numbers_array:
    print("10 is Present")
else:
    print("10 is not Present")
```

- <b>Raise an alarm if number 15 is NOT present in the array. Accorgingly, print "15 NOT present!", else "All is fine."


```python
numbers_array = [1, 5, 10, 15, 20]
if 15 not in numbers_array:
    print("15 is not present")
else:
    print("All is Fine")
```

```python
numbers_array = [1, 5, 10, 15, 20]
if 15 in numbers_array:
    print("All is fine")
else:
    print("15 is not present")
```

- <b> Check if the number contained in the last array position is greater than the number contained in the first array position. Print "Yes™ if true, else do not print anything.

```python
numbers_array = [1, 5, 10, 15, 20]

if numbers_array[4]> numbers_array[0]:
    print("Yes last array position is greater than the number containes in the first array")
else:
    print("")
```

```python
numbers_array = [1, 5, 10, 15, 20]

if numbers_array[len(numbers_array)-1]> numbers_array[0]:
    print("Yes last array position is greater than the number containes in the first array")
else:
    print("")
```

- <b> Check if the sum of the first 3 numbers in the list is greater than, equal to or less than the last number in the given list. Accordingly, print
    "Greater Than", "Equal to" or "Less Than" - based on the result of the comparisons.

```python
numbers_array = [1, 5, 10, 15, 20]
if(numbers_array[0]+numbers_array[1]+numbers_array[2]) > numbers_array[len(numbers_array)-1]:
    print("Greater Than")
elif(numbers_array[0]+numbers_array[1]+numbers_array[2])==numbers_array[len(numbers_array)-1]:
    print("Equal To")
else:
    print("Less Than")
```
