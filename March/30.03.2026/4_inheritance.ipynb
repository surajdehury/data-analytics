```python
class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print(f"{self.name} says",end=':')
```

```python
class Dog(Animal):
    def speak(self):
        super().speak()
        print("Woof")
```

```python
class Cat(Animal):
    def speak(self):
        super().speak()
        print("Meow")
```

```python
dog=Dog('Rufus')
cat=Cat('Shiny')
```

```python
dog.name , cat.name
```

```python
dog.speak()
```

```python
cat.speak()
```
