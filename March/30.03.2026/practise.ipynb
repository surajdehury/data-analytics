```python
#scope#calculate area and circumference of a circle
PI=3.414
def calculate_area(r):
    global PI
    area=PI*(r**2)
    return area
def calculate_circumference(r):
    global PI
    circumference=2*PI*r
    return circumference
```

```python
calculate_area(5)
```

```python
class Person:
    def __init__(self,name,age,gender,speed):
        self.name=name
        self.age=age
        self.gender=gender
        self.speed=speed
    def speak(self):
        print(f"Hello ,my name is {self.name} and I'm {self.age} years old.")
    def walk(self):
        print(f"I am walking with a speed of {self.speed}")
p1=Person('Suraj',24,'Male',34)
p2=Person('Mia',34,'Female',25)
```

```python
p1.speak()
```
