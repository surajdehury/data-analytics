# Scope 

```python
#Global Scope-variable that are defined outside of any function have global scope
#Local Scope-variable that are defined inside a function have local scope
```

## Global Scope

```python
x=5
def show():
    print(x)
print(x)
```

```python
show()
```

## Local Scope

```python
x=5 #Global scope#this can used inside as well as outside
def show():
    y=10#Local Scope #This can only be used inside cannot be used outside
    print(f'x:{x}')
    print(f'y:{y}')
show()
```

### Calculate the area and circumference of a circle

```python
PI=3.14159
def calculate_area(r): #r=radius
    global PI
    area=PI*(r**2)
    return area
def calculate_circumference(r):#circumference
    global PI
    circumference=2*PI*r
    return circumference
```

```python
calculate_area(5)
```

```python
calculate_circumference(5)
```
