# Class

### Creating a class

```python
class Person:
    ##attributes
    def __init__(self,name,age,gender,speed):
        self.name=name
        self.age=age
        self.gender=gender
        self.speed=speed
    ##method
    def speak(self):
        print(f"Hello ,my name is {self.name} and I'm {self.age} years old.")

    def walk(self,speed):
        print(f"I am walking with a speed of {self.speed}")
```

## Object

```python
p1=Person('Suraj',24,'Male',35)
p2=Person('Mia',34,'Female',25)
```
