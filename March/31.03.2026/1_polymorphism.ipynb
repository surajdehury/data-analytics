# Polymorphism

```python
class Shape:
    def area(self):
        pass
```

```python
class Rectangle(Shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width
```

```python
PI=3.14
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        global PI
        return PI*self.radius**2
```

```python
rect=Rectangle(10,2)
```

```python
circle=Circle(5)
```

```python
rect.area()
```

```python
circle.area()
```
