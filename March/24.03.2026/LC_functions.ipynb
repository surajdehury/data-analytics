```python
#Function without argument
def greet_user():
    for i in range(1,4):
        print("Hello Welcome to put program")
```

```python
greet_user()
```

```python
#Function with argument
def greet_user(n):
    for i in range(1,n):
        print("Hello Welcome to put program")
```

```python
#calling the fucntion
greet_user(5)
```
