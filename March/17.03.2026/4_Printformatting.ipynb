## **Print Formating** ##

### **Concatenation** ###

```python
name="Suraj"
print("Hello," + name + "!")
```

### ***2.String Interpretation*** ###

```python
fname="Suraj" 
lname="Dehury"
print("Hello,{} {}!".format(fname,lname))
```

### 3.String Formatting with %

```python
#%s-string
#%d-Integers
#%f-Flaoting point numbers
#%x-Hexadecimal
```

```python
fname="Suraj" 
lname="Dehury"
age=18

print("Hello %s %s!You are %d years old"%(fname,lname,age))
```

```python
revenue=899
print("The revenue is %0.2f" %(revenue))
```

### f-string

```python
print(f"Hello {fname} {lname}.You are {age} years old")
```

### 5.Specifying the separator and End Character:
