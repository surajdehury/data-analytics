## My First Python Program

```python
student1=('Raju',8,'Class-3')
student2=('Ramesh',8,'Class-3')
student3=('Raja',7,'Class-2')
student4=('Romeo',9,'Class-4')
student5=('Mallu',10,'Class-5')

students=[student1,student2,student3,student4,student5]

details={'name':0,'age':1,'class':2}#dictionary,metadata

student_rank=int(input("Enter the Rank of student for details:"))

index=student_rank-1

print(f"Name:{students[index][details['name']]}")
print(f"Age:{students[index][details['age']]}")
print(f"Class:{students[index][details['class']]}")

```
