#### 2. MAP

**Question 2**: It was decided that all the aquarium creatures are in fact going to move into the same tank. Update the records to reflect that all of the creatures are moving into tank `42`


```python
aquarium_creatures = [
    {"name": "sammy", "species": "shark", "tank number": 11, "type": "fish"},
    {"name": "ashley", "species": "crab", "tank number": 25, "type": "shellfish"},
    {"name": "jo", "species": "guppy", "tank number": 18, "type": "fish"},
    {"name": "jackie", "species": "lobster", "tank number": 21, "type": "shellfish"},
    {"name": "charlie", "species": "clownfish", "tank number": 12, "type": "fish"},
    {"name": "olly", "species": "green turtle", "tank number": 34, "type": "turtle"}
]
```

```python
def change_tank(input_, new_tank_number= 42):
    input_["tank number"] = new_tank_number
    return input_
```

**with loop**

```python
aquarium_creatures_new = []
for creature in aquarium_creatures:
    new_creature = change_tank(creature)
    aquarium_creatures_new.append(new_creature)
```

```python
aquarium_creatures_new
```

**with map**

```python
aquarium_creatures_new = list(map(change_tank, aquarium_creatures))
```

```python
aquarium_creatures_new
```

`Observation`: A multiple lines code for LOOP is re-written in a single line!

#### 3. MAP & Lambda

**Question 3**: convert each word into lower and upper cases along with getting the length of each word for below sentence

```python

sentence = "Let us all just take a moment to breathe, please!"
```

```python
all_words_info = []
for word in  sentence.split():
    all_words_info.append([word.upper(), word.lower(), len(word)])
```

```python
all_words_info
```

```python
words = sentence.split()
results = map(lambda w: [w.upper(), w.lower(), len(w)], words)
list(results)
```

#### 4. Filtering

**Question 4**: In the below list of countries - filter out those without `land` in the names

```python
countries= ['Germany', 'Finland', 'Ireland', 'India', 'United Kingdom', 'United States', 'Turkey', 'Poland']
```

```python
land_countries = []
for country in countries:
    if "land" in country:
        print(country)
        land_countries.append(country)
```

**with filter and lambda**

```python
list(filter(lambda x: 'land' in x.lower(), countries))
```

`Observation`: Conditional Operations and loops are replaced by single line of code with filter

**Question 6** : Below is the sports data for students

Get which batch the students belong to using MAP based on-
  1. age <= 18 : B1
  2. age > 18 and <=22 : B2
  3. age > 22 : B3

```python
students = [
        { 'id': "001", 'name': "Anish", 'sports': "Cricket", 'age': 22 },
        { 'id': "002", 'name': "Smriti",'sports': "Basketball" , 'age': 17},
        { 'id': "003", 'name': "Rahul", 'sports': "Cricket" , 'age': 24},
        { 'id': "004", 'name': "Bakul", 'sports': "Basketball", 'age': 18 },
        { 'id': "005", 'name': "Nikita", 'sports': "Hockey", 'age':16 }
]
```

```python
def age_group(age):
    if age <= 18:
        return "B1"
    elif age> 18 and age<=22:
        return "B2"
    else:
        return "B3"
    
```

```python
list(map(lambda x: age_group(x['age']), students))
```

**Question 7** :get students playing basketball only

```python
list(filter(lambda x: x['sports'].lower()== 'basketball', students))
```

**Conclusions**:
  1. Filter and Map saved lot of looping process and in general reduces the code size and complexities in code
  2. Map and lambda should be used where processing- "mapping" needs to be performed
  3. Filter and lambda should be used for cases where records are to be filtered based on certain conditions
