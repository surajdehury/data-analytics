```python
n=int(input("Enter the number to find the factorial:"))
def myfact(n):
    if n==0:
        return 1
    else:
        return n*myfact(n-1)
result=myfact(n)
print("Factorial of",n,"is:",result)
```

```python
#wap in python with recursive function ,to add all numbers between 1 and n
n=int(input("Enter the number:"))
def naturaladdition(n):
    if n==0:
        return 0
    else:
        return n+naturaladdition(n-1)
result=naturaladdition(n)
print("Addition of",n,"is:",result)
```

```python
def calculate_square(x):
    return x*x

num=int(input("Enter the number:"))
result=calculate_square(num)
print("Square of ",num,"is:",result)
```

```python
square=lambda x:x*x
num=int(input("Enter the number:"))
result=square(num)
print("Square of ",num,"is:",result)
```

```python
multiply=lambda x,y:x*y
n=int(input("Enter the number x:"))
m=int(input("Enter the number y:"))
result=multiply(n,m)
print("Multiplication of",n,"and",m,"is:",result)
```
