# Looping through dictionaries

```python
person={"first_name":"Suraj","last_name":"Dehury","age":24,"gender":"Male"}
```

```python
for key in person:
    print(key,":",person[key])
```

```python
for key,value in person.items():
    print(key,":",value)
```

```python
for key in person.keys():
    print(key,":",person[key])
```

```python
#all the value
for value in person.values():
    print(value)
```
