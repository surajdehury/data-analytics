```python
#Palindrome
s=input()
s=s.lower()
rev=""
for i in range(len(s)-1,-1,-1):
    rev+=s[i]
if s== rev:
    print("Palindrome")
else:
    print("Not Plaindrome")
```

```python
# reverse a string
s=input()
rev=""
for i in range(len(s)-1,-1,-1):
    rev+=s[i]
print(rev)
```

```python
n = int(input())
arr = list(map(int, input().split()))

res = [x for x in arr if x != 0]
res += [0] * (n - len(res))

print(*res)
```

```python
n = int(input())
arr = list(map(int, input().split()))

first = second = -10**9

for num in arr:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print(second)
```

```python
n=int(input())
total=0
disc=0
if n<=1000:
    disc=(5*n)//100
    total=n-disc
elif 1000<n<=5000:
    disc=(10*n)//100
    total=n-disc
else:
    disc=(15*n)//100
    total=n-disc
n=total
print(n)
```

```python
# WAP to print n no inputted by user
n=int(input("Enter the prime numbers to be printed"))
count=1
i=2
while count<=n:
    prime=True
    for j in range (2,i):
        if i%j==0:
            prime=False
            break
    if prime==True:
        print(i, " ")
        count+=1
    i+=1
```
