# While Loops

```python
count=0
while count<5:
    print(count)
    count+=1
```

- <b>Example 1: Printing the squares of numbers in a range

```python
start=int(input("Enter the starting number:"))
end=int(input("Enter the ending number:"))
number=start
while number<=end:
    print("The square of the number is:",number**2)
    number+=1
```
