# Reduce

```python
numbers=[1,2,3,4,5]
def add(x,y):
    return x+y
```

```python
from functools import reduce
summation=reduce(add,numbers)
summation
```

```python
numbers=[3,2,4,5,8]
def multiply(a,b):
    return a*b
from functools import reduce
multiplication=reduce(multiply,numbers)
multiplication
```
