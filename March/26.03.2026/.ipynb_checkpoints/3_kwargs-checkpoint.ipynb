# KWARGS

```python
def print_kwargs(**kwargs):
    print(type(kwargs))
    return kwargs
```

```python
print_kwargs(a=1,c=3,d='string')
```

```python
def print_kwargs(**kwargs):
    for key,value in kwargs.items():
        print(f'{key}={value}')
```

```python
print_kwargs(a=1,b=2,c=3)
```

### Example 1:

```python
def calculate_area(shape,**kwargs):
    if shape=='circle':
        return 3.14*kwargs['radius']**2
    elif shape=='rectangle':
        return kwargs['width']*kwargs['height']
    elif shape =='triangle':
        return 0.5*kwargs['base']*kwargs['height']
    else:
        raise ValueError("Invalid shape")
            
```

```python
calculate_area('circle',radius=5)
```

```python
area=calculate_area('rectangle',width=4,height=5)
area
```

```python
calculate_area("triangle",base=5,height=7)
```
