## Function Example 1

- <b> Create a function that returns the current time

```python
import datetime
now=datetime.datetime.now()
print(now)
now.strftime("%H:%M:%S")
```

```python
def get_currtime():
    import datetime
    now=datetime.datetime.now()
    time=now.strftime("%H:%M:%S")
    return time
```

```python
get_currtime()
```

## Function with Parameters

```python
def add(a,b):
    return a+b
```

```python
add(5,7)
```

```python
def cal_avg(data:list):
    total=sum(data)
    avg=total/len(data)
    return avg
```

```python
data=[3,4,8,1,7,6,5]
```

```python
cal_avg(data)
```

## Functions with args and kwargs

```python
def sum(*args):
    total=0
    for arg in args:
        total=total+arg
    return total
```

```python
sum(5,7)
```

```python
sum(1)
```

```python
sum(2,3,4,5,6,7,8,9,10)
```

```python
#avg of given number
def average(*numbers):
    return sum(*numbers)/len(numbers)
```

```python
average(1,2,3,4,5)
```
