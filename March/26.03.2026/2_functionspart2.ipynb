## Functions with default parameters

```python
def divide(a,b=1):
    return a/b
```

```python
divide(5)
```

```python
divide(5,2)
```

## Calculate the total cost after applying tax_rate

```python
def cal_total_cost(cost,tax_rate=0.1):
    total_cost=cost+ cost*tax_rate
    return total_cost
```

```python
cal_total_cost(100)
```

```python
cal_total_cost(100,0.155)
```

## Function returning multiple values

```python
#returning multiple values
def divide(a,b):
    quotient=a//b
    remainder=a%b
    return quotient,remainder
divide(18,3)
```

```python
#calcute stack of given data
data=[1,2,3,4,5]
def calculate_stats(data:list):
    total=sum(data)
    count=len(data)
    mean=total/count
    return total,count,mean
```

```python
total,count,mean=calculate_stats(data)
```

```python
print("Total:",total)
print("Count:",count)
print("Mean:",mean)
```
