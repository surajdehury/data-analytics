# Introduction to Loops

```python
numbers=[1,2,3,4,5]
```

```python
sum_num=0
for number in numbers:
    sum_num+=number
print("The sum of numbers is:",sum_num)
```

```python
sum_num=0
for x in numbers:
    sum_num+=x
print("The sum of numbers is:",sum_num)
```

```python
numbers=[1,2,3,4,5]
sum_num=0
for i in range (len(numbers)):
    sum_num+=numbers[i]
    print(numbers[i])
print("The sum of numbers is:",sum_num)
```

```python
#index 2 to len(numbers)
numbers=[1,2,3,4,5]
sum_num=0
for i in range (2,len(numbers)):
    sum_num+=numbers[i]
    print(numbers[i])
print("The sum of numbers is:",sum_num)
```

```python
#sum of first 2 elements
numbers=[1,2,3,4,5]
sum_num=0
for i in range (2):
    sum_num+=numbers[i]
    print(numbers[i])
print("The sum of numbers is:",sum_num)
```

```python
numbers=[1,2,3,4,5]
sum_num=0
i=0
while i<len(numbers):
    sum_num+=numbers[i]
    i+=1
print("The sum of numbers is:",sum_num)
```
