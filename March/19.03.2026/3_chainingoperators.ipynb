# Chaining operators

- <b> User to input any number.If the number is between 10 and 20,then print that "Valid Entry",else "Invalid Request"

```python
user_input=int(input("Enter a number:"))
if (user_input>10) and (user_input<20):
    print("Vaild Entry")
else:
    print("Invalid Entry")
```
