```python
x=int(input("Enter the vale of x:"))
y=int(input("Enter the vale of y:"))

if x>y:
    print('x is greater than y')
else:
    if x==y:
        print('x and y are equal')
    else:
        print('y is greater than x')
```

```python
#else if statement(elif)
x=int(input("Enter the vale of x:"))
y=int(input("Enter the vale of y:"))

if x>y:
    print('x is greater than y')
elif x==y:
    print('x and y are equal')
else:
    print('y is greater than x')
```
