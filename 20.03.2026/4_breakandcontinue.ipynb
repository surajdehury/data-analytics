# Break And Continue

```python
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
i=0
while i <10:
    if i == 5:
        break
    print(i)
    i+=1
```

#### continue

```python
for i in range(10):
    if i % 2 ==0:
        continue
    print(i)
```

```python
i=0
while i<10:
    i+=1
    if i%2==0:
        continue
    print(i)
```

- <b> Practical Example 1a:

- <b>We want to find the first customer who spent more than 1000 rupees.

```python
purchases=[
    {'customer':'John','amount':900},
    {'customer':'Jane','amount':1200},
    {'customer':'Jim','amount':800},
    {'customer':'Jack','amount':1500},
    {'customer':'Jill','amount':30},
    {'customer':'Josh','amount':700},
    {'customer':'Jet','amount':10}
]
for element in purchases:
    if element['amount'] >1000:
        print("First customer who spent over :",element['customer'])
        break
        
```

- <b>Practical Example 1b:

- <b> Find out the total number of purchases across akk customers.However while calculating the sum,ignore purchases that are less than 100 rupees

```python
total_spent=0
for purchase in purchases:
    if purchase['amount']<100:
        continue
    total_spent+=purchase['amount']
print('Toal spent by customers who spent over',total_spent)
```

```python
total_spent=0
for purchase in purchases:
    if purchase['amount']>=100:
        total_spent+=purchase['amount']
print('Toal spent by customers who spent over',total_spent)
```

- <b> Write a python program that prints -'n' being input by the user 

```python
n=int(input("Enter the number of prime numbers to be printed:"))
count=1
i=2
while count<=n:
    prime=True
    for j in range(2,i):
        if i%j==0:
            prime=False
            break
    if prime==True:
        print(i," ")
        count+=1
    i+=1
    
```
