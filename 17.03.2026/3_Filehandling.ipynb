### **File Handling** ###

File handling is basically a crucial aspect of any programming language and is used to read,write and manipulate files.

```python
import os
os.getcwd()
```

### **1.Opening a File** ###

```python
"""r=read mode,used to read an existing file
w=write mode,used to create a new file or overwrite an existing file
a=append mode,used to add new data to an existing file
x=exclusive creation mode,used to create a new file but fail if the file already exists"""
```

```python
file=open('ok.txt','r')
```

```python
type(file)
```

### **2.Reading a File** ###

```python
txt=file.read()
```

```python
txt
```

### **3.Writing a File** ###

```python
file=open('ok.txt','w')
```

```python
file.write("Hello Again,This is me\n")
```

```python
file.close()
```

### **4.Appending a File** ###

```python
file =open("ok.txt",'a')
```

```python
file.write("This is the 2nd line in the file.\n")
```

```python
file.close()
```

### **5.Exclusive Creation mode** ###

```python
file=open('testforexclusive.txt','x')
```

```python
file.write("This is a message for the new file")
```

```python
file.close()
```
