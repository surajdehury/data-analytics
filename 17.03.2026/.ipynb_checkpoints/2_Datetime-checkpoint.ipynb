```python
import datetime
now=datetime.datetime.now()
print(now)
```

```python
type(now)
```

### **2.Date Formatting**

```python
now.strftime("%d %b %Y")
```

```python
now.strftime("%d %B %Y")
```

### **3. Creating a Datetime Object from a string**

```python
date_string="2022-03-10 12:30:45"
```

```python
parsed_date=datetime.datetime.strptime(date_string,"%Y-%m-%d %H:%M:%S")
```

```python
print(parsed_date)
```

```python
type(parsed_date)
```

### **5.Performing Operations with datetime objects**

```python
import datetime
```

```python
now=datetime.datetime.now()
```

```python
one_day=datetime.timedelta(days=1)
```

```python
print(now)
```

```python
print(now+one_day)
```

```python
print(now-one_day*3)
```

```python
one_hour=datetime.timedelta(hours=1)
```

```python
type(one_hour)
```

```python
print(now-one_hour)
```

```python
print(now+one_hour*2)
```

```python
one_hour
```

```python
one_minute=datetime.timedelta(minutes=1)
one_second=datetime.timedelta(seconds=1)
one_millisecond=datetime.timedelta(milliseconds=1)
one_microsecond=datetime.timedelta(microseconds=1)
```

### **6.Finding Date Difference** ###

```python
date1=datetime.datetime(2022,3,1,12,0,0)
date2=datetime.datetime(2022,2,28,12,0,0)
```

```python
print(date1)
print(date2)
```

```python
diff=date1- date2
diff
```

```python
print(diff)
```
