# List Comprehension

```python
numbers=[1,2,3,4,5]
squared_numbers=[]
squared_numbers={num:num**2 for num in numbers}
print(squared_numbers)
```

```python
numbers=[1,2,3,4,5]
even_numbers={num:True if num%2==0 else False for num in numbers}
print(even_numbers)
```

```python
numbers=[1,2,3,4,5]
odd_numbers={num:False if num%2==0 else True for num in numbers}
print(odd_numbers)
```
