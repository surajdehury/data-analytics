## Practise

```python
import pandas as pd
```

```python
data=pd.read_csv('data_clean.csv')
```

```python
data.head()
```

```python
data=data.drop('Unnamed: 0',axis=1)
```

```python
data.head()
```

```python
data=data.drop('Temp C',axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Month'].isnull().sum()
```

```python
data['Month']=data['Month'].fillna(5)
```

```python
#changing temp to numeric
data['Temp']=pd.to_numeric(data['Temp'],errors='coerce')
```

```python
data.isnull().sum()
```

```python
#changing numeric to categories
data['Weather']=data['Weather'].astype('category')
```

```python
#checking of duplicated data
data[data.duplicated()].shape
```

```python
#check the duplicated data
data[data.duplicated()]
```

```python
#Droping of the duplciated datas from table and storing it in a new place
data1=data.drop_duplicates()
```

```python
#final check for duplicacy
data1[data1.duplicated()]
```

```python
#Visualizing the data
```

```python
import seaborn as sns
cols=data1.columns
colours=['blue','yellow']
sns.heatmap(data1[cols].isnull(),
           cmap=sns.color_palette(colours))
```

```python
#renaming the coloums
data2=data1.rename({'Solar.R':'Solar'},axis=1)
```

```python
data2
```

```python
#mean Imputation 
n=data2['Ozone'].mean()
n
```

```python
data2['Ozone']=data2['Ozone'].fillna(n)
print(data2)
```

```python
data2.isnull().sum()
```

```python
m=data2['Solar'].mean()
print(m)
```

```python
data2['Solar']=data2['Solar'].fillna(m)
```

```python
print(data2)
```

```python
#for categorical 
o=data2['Weather'].mode().iloc[0]
print(o)
```

```python
data2['Weather']=data2['Weather'].fillna(o)
print(data2.head())
```

```python
data2.isnull().sum()
```

```python
#outlier detection
data2['Ozone'].hist()
```
