### Create data Frame

```python
data1={'RollNo':[1,2,3],
    'Subject':['A','B','C']}
```

```python
df1=pd.DataFrame(data1)
```

```python
df1
```

```python
data2={'RollNo':[4,5,6],
      'Subject':['D','E','F']}
```

```python
df2=pd.DataFrame(data2)
```

```python
df2
```

```python
result_df=pd.concat([df1,df2],axis=0,ignore_index=True)
print(result_df)
```

```python
df3=pd.DataFrame({'ID':[1,2,3],'Name':['Alice','Bob','Charlie']})
df4=pd.DataFrame({'ID':[2,3,4],'Age':[25,30,35]})
```

```python
merge_inner=pd.merge(df3,df4,on='ID',how='inner')
```

```python
print('\n step 2:Inner Join(Default')
print(df3)
```

```python
merge_outer=pd.merge(df3,df4,on='ID',how='outer')
print("\n Step 3:Outer Join")
print(merge_outer)
```

```python
merge_left=pd.merge(df3,df4,on='ID',how='left')
print("\n step 4: Left Join")
print(merge_left)
```

```python
merge_right=pd.merge(df3,df4,on='ID',how='right')
print("\n step 5:Right Join")
print(merge_right)
```
