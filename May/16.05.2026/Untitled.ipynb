```python
import pandas as pd
data=pd.read_csv('data_clean.csv')
```

```python
data.head(5)
```

```python
data_original=data
```

```python
data=data.drop('Unnamed: 0',axis=1)
```

```python
data.head(1)
```

```python
data=data.drop('Temp C',axis=1)
```

```python
data.head(1)
```

```python
data=data.rename({'Solar.R':'Solar'},axis=1)
```

```python
data.head(1)
```

```python
data.isnull().sum()
```

```python
data['Temp']=pd.to_numeric(data['Temp'],errors='coerce')
```

```python
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
```

```python
data['Month']=data['Month'].fillna(5)
```

```python
data['Weather']=data['Weather'].astype('category')
```

```python
data.isnull().sum()
```

```python
data.head(10)
```

```python
n=data['Ozone'].mean()
data['Ozone']=data['Ozone'].fillna(n)
data['Ozone'].isnull().sum()
data.head(7)
```

```python
n=data['Solar'].mean()
data['Solar']=data['Solar'].fillna(n)
data.head(7)
```

```python
data1=pd.get_dummies(data,columns=['Weather'],dtype=int)
```

```python
data1.head(5)
```
