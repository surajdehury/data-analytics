# Concate,combining data frame

```python
import pandas as pd
```

```python
data1={'Roll No':[1,2,3],
      'Subject':['A','B','C']}
df1=pd.DataFrame(data1)
```

```python
data1
```

```python
df1
```

```python
data2={'Roll No':[4,5,6],
      'Subject':['D','E','F']}
df2=pd.DataFrame(data2)
```

```python
df2
```

```python
#concatenate dataframe along with column and index remains as it is
result_df=pd.concat([df1,df2],axis=0,ignore_index=True)
print(result_df)
```

# Merging

```python
import pandas as pd
df3=pd.DataFrame({'ID':[1,2,3],'Name':['Alice','Bob','Charlie']})
df4=pd.DataFrame({'ID':[2,3,4],'Age':[25,30,35]})
```

```python
df3.head()
```

```python
df4.head()
```

```python
merge_inner=pd.merge(df3,df4,on='ID',how='inner')
print("\nStep2: Inner Join (default)")
print(merge_inner)
```

```python
merge_left=pd.merge(df3,df4,on='ID',how='left')
print("\n Step 3:Left Join")
print(merge_left)
```

```python
merge_right=pd.merge(df3,df4,on='ID',how='right')
print("\n Step 3:Right Join")
print(merge_right)
```

```python
merge_outer=pd.merge(df3,df4,on='ID',how='outer')
print("\n Step 5: Outer Join")
print(merge_outer)
```

# Join 

```python
import pandas as pd
df5=pd.DataFrame({'Roll No':[1,2,3],'Name':['x','y','z']},index=[1,2,3])
df6=pd.DataFrame({'Score':[4,5,6],'Grade':['a','b','c']},index=[2,3,4])
```

```python
df5
```

```python
df6
```

```python
resultjoin=df5.join(df6,how='inner')
print(resultjoin)
```
