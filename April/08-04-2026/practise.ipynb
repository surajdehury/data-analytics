```python
def square():
    try:
        num=int(input("Enter the number:"))
        return num**2
    except ValueError:
        print("Invalid Input,Please enter a input variable")
```

```python
def divide(num,den):
    try:
        return float(num)/float(den)
    except ZeroDivisionError:
        print("Cannot divide by zero,the denominator should be non-zero")
    except Exception as e:
        print("An unknown error occured:",e)
```

```python
divide(8,2)
```

```python
divide(8,0)
```

```python
divide(8,'jj')
```
