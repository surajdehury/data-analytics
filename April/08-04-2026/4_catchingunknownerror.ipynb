```python
#Example
def divide(num,den):
    try:
        return float(num)/float(den)
    except ZeroDivisionError:
        print("Cannot divide by zero,the denominator should be non-zero")
    except Exception as e:
        print("An unknown error occured:",e)
```

```python
divide(8,2)
```
