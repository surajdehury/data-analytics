```python
#nested
def outer_function():
    def inner_function():
        print("Inner One")
    print("Outer One")
    inner_function()
```

```python
outer_function()
```

```python
def add(a,b):
    def check_input():
        if not isinstance(a,(int,float)):
            raise TypeError("a must be a number")
        if not isinstance(b,(int,float)):
            raise TypeError("b must be a number")
    check_input()
    return a+b
```

```python
add(8,2)
```

```python
add(8,'j')
```
