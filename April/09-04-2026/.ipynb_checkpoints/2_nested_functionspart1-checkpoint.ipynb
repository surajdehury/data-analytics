# Nested functions

```python
#Nested functions also known as inner functions are functions that are defined inside other functions
```

```python
"""Benefits:
> Encapsulation(hiding functions from external access) basically defined in main which is not visible to outer world
>Inner helper function
>Code decoder functions to add behaviour to existing functions"""
```

```python
def outer_function():
    def inner_function():
        print("Hello from inner function")
    print("Hello from outer function")
    #inner_function()
```

```python
outer_function()
```

```python
inner_function()
```

```python
def outer_function():
    def inner_function():
        print("Hello from inner function")
    print("Hello from outer function")
    inner_function()
```

```python
outer_function()
```
