```python
def add_numbers(a,b):
    def check_input(a,b):
        if not isinstance(a,(int,float)):
            raise TypeError("a must be a number")
        if not isinstance(b,(int,float)):
            raise TypeError("b must be a number")
    check_input(a,b)
    return a+b
```

```python
add_numbers(3,5)
```

```python
add_numbers(3,'jj')
```
