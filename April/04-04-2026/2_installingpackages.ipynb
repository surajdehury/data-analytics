# 3rd Party Packages

- it refers to a package or library that is developed and maintained by someone other than creators of the python programming language or the standard library

- helps us to add functionality or solve a specific problems,there are thousands or 3rd party packages covering a wide variety of usecases. 

- <b>Some are:
        Numpy-for numerical computing
        pandas-for data analysis
        matplotlib-for dara visualization
        django-for web development
        flask-for web application development
        requests-for https requests

## Different ways to install packages in python

1) using pip
-pip install <package_name>
2) using anaconda
-open conda comand prompt
-and write
-conda install <package_name>
3) using conda forge
-conda install -c conda-forge <package_name>
