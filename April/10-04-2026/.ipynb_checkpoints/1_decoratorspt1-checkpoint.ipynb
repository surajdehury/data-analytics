# Decorators

## Creating a Decorator

```python
def my_decorator(func):
    def wrapper():
        print("statements before the function is called")
        func()
        print("statements after the function is called")
    return wrapper
```

```python
@my_decorator
def my_function():
    print("statement inside the function")
```

```python
my_function()
```

```python
my_function()
```
