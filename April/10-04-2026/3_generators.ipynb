# Gererators

- it is a special typ eof function that returns an iterator object.Instead of returning all the values at once like a regular function,a generator yiels one value at a time using the yieeld keyword

```python
def count_up_to(n):
    i=1
    while i<=n:
        yield i
        i+=1
```

```python
gen=count_up_to(10)
```

```python
next(gen)
```
