- <b>The init method

```python
class Person:
    def __init__ (self,name,age):
        self.name=name
        self.age=age
```

```python
p1=Person('John',15)
```

```python
p1.age,p1.name
```

- <b> Str method

```python
print(p1)
```

```python
class Person:
    def __init__ (self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name} ({self.age}years old)"
```

```python
p1=Person("John",15)
```

```python
print(p1)
```

-<b> the repr method-used to recreate the object

```python
class Person:
    def __init__ (self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name} ({self.age}years old)"
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
```

```python
p1=Person('John',15)
```
