```python
#eq method
class Person:
    def __init__ (self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name} is {self.age} hyears old."
    def __repr__(self):
        return f"Person(name={self.name},age={self.age})"
    def __eq__(self,other):
        return self.name==other.name and self.age==other.age
```

```python
p1=Person("John",15)
```

```python
p2=Person("John",15)
```

```python
p3=Person("Johnny",20)
```

```python
p1==p2
```

```python
p1==p3
```

```python
#add
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return (self.x+other.x,self.y+other.y)
```

```python
pt1=Point(10,30)
```

```python
pt2=Point(20,30)
```

```python
pt1+pt2
```
