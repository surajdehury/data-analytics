# Comparison Shortcuts

- <b> Check if the number contained in the last array position is greater than the number contained in the first array position.Print"Yes" if true ,else do not print anything

```python
numbers_array=[1,5,10,15,20]
if(numbers_array[len(numbers_array)-1] > numbers_array[0]):
    print("Yes")
else:
    print(" ")
```

- <b> Check if the sum of the first 3 numbers in the list is greater than, equal to or less than the last number in the given list. Accordingly, print "Greater Than", "Equal to" or "Less Than" - based on the result of the comparisons.

```python
numbers_array = [1, 5, 10, 15, 20]
if(numbers_array[0]+numbers_array[1]+numbers_array[2]) > numbers_array[len(numbers_array)-1]:
    print("Greater Than")
elif(numbers_array[0]+numbers_array[1]+numbers_array[2])==numbers_array[len(numbers_array)-1]:
    print("Equal To")
else:
    print("Less Than")
```
