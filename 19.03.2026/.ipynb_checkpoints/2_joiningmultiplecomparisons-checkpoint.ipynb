# Joining Multiple comparisons using logical operators

- <b> User to input any number.If the number is between 10 and 20,then print that "Valid Entry",else "Invalid Request"

```python
input_number=int(input("Enter a number:"))
if (input_number>10) and (input_number<20):
    print ("Valid Entry")
else:
    print("Not Valid Request")
```

- <b> User to input any number.This time,if the number is less than 10 or greater than 20,then print"Okay" else print"Danger"

```python
input_number=int(input("Enter a number:"))
if (input_number< 10) or (input_number>20):
    print("Okay")
else:
    print("Danger")
```

- <b> Request Request the User to input the Gender, Age and Whether Covid Symptoms are Present for a patient. Accordingly, the output should suggest based on the following: If Covid Symptoms present, refer to a doctor only under the following conditions:
• Gender: Male and Age > 50 years
• Gender: Female and Age > 60 years
• Age < 8 years
Else, if the patient has Covid symptoms but the above conditions are not met, then recommend monitoring at home and visit a doctor if conditions worsen. If no Covid symptoms are present, ask the Patient to Exercise Caution and Stay Safe.

```python
gender=input("Enter Gender of Patient(M/F):")
age=int(input("Enter Age of Patient:"))
covid_symptoms=input("Does the Patient have COVID Symptoms(Y/N):")
if (covid_symptoms=='Y') and ((gender=='M' and age>50) or (gender=='F' and age>60) or(age<8)):
    print('Refer to a Doctor.')
elif(covid_symptoms=='Y'):
    print('Recommend Monitoring at home,if condition worsen visit a dotor.')
else:
    print('Exercise caution and Stay Safe')
```
